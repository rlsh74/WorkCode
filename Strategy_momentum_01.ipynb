Universe: DJIA stocks

1. Check the performance over every single stock over the last 12 month (t-12)
2. Skip the most recent month (we have a 11 month window)
3. Buy the winners (top quintile) and short the losers
4. Hold the portfolio 1 month

Then we compare the return to the benchmark (DJIA)

```python
import pandas as pd
import numpy as np
import pandas_datareader.data as reader
import datetime as dt
from dateutil.relativedelta import relativedelta
```

```python
#Get the tickers 

table = pd.read_html('https://en.wikipedia.org/wiki/Dow_Jones_Industrial_Average')[1]
table
```

```python
tickers = table.Symbol.tolist()
tickers
```

```python
#Get the prices
start = dt.datetime(2018, 1, 31)
end = dt.datetime(2020, 1, 31)

df = reader.get_data_yahoo(tickers, start, end) ['Adj Close']
```

```python
df
```

```python
#Calculate monthly returns by calculating daily returns

mtl_ret = df.pct_change().resample('M').agg(lambda x: (x + 1).prod() - 1)
mtl_ret
```

```python
#Calculate returns over the past 11 months

past_11 = (mtl_ret + 1).rolling(11).apply(np.prod) - 1
past_11
```

```python
# Define formation date for our portfolio

formation = dt.datetime(2019, 12, 31)
end_measurement = formation - relativedelta(months=1)
```

```python
formation
```

```python
end_measurement
```

```python
ret_12 = past_11.loc[end_measurement]
ret_12
```

```python
ret_12 = ret_12.reset_index()
ret_12
```

```python
#Calculate the qintiles (make distributions of our returns to 5 equal parts). The hiest quintile is number 4, 
#and the lowerst has number 0

ret_12['Quintile'] = pd.qcut(ret_12.iloc[:, 1], 5, labels=False)
ret_12
```

```python
winners = ret_12[ret_12.Quintile == 4]
winners
```

```python
losers = ret_12[ret_12.Quintile == 0]
losers
```

```python
# We buy winners and short losers
# Get the monthly portfolio return

winnerret = mtl_ret.loc[formation + relativedelta(months=1), df.columns.isin(winners.Symbols)]
loserret = mtl_ret.loc[formation + relativedelta(months=1), df.columns.isin(losers.Symbols)]
```

```python
winnerret
```

```python
loserret
```

```python
Momentumprofit = winnerret.mean() - loserret.mean()
Momentumprofit
```

```python
#Lets check the performance of our benchmark
DJI = reader.get_data_yahoo('^DJI', start, end)['Adj Close'].pct_change().resample('M').agg(lambda x: (x+1).prod() -1)
```

```python
DJI
```

So, we outperformed our benchmark by 0.0858 + 0.0099 = 0.0957 or 9.57%!
